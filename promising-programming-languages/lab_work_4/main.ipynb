```python
import random
import math

class Factory:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def geolocation(self):
        print(f"Координаты: x={self.x}, y={self.y}")

class Employee(Factory):
    def __init__(self, x, y):
        super().__init__(x, y)

    def ChangePosition(self, dx, dy):
        self.x += dx
        self.y += dy
        print(f"Работник переместился. Новые координаты: x={self.x}, y={self.y}")

    def department(self):
        print("Работник принадлежит к отделу продаж.")

factory = Factory(5, 10)
print("Информация о заводе:")
factory.geolocation()
print("-" * 20)

employee = Employee(50, 50)
print("Начальная информация о работнике:")
employee.geolocation()
employee.department()
print("-" * 20)

random_dx = random.randint(-20, 20)
random_dy = random.randint(-20, 20)
print(f"Смещение работника на dx={random_dx}, dy={random_dy}...")
employee.ChangePosition(random_dx, random_dy)
print("-" * 20)

distance = math.sqrt((factory.x - employee.x)**2 + (factory.y - employee.y)**2)

print("Итоговая информация:")
print("Координаты завода:")
factory.geolocation()
print("Текущие координаты рабочего:")
employee.geolocation()
print(f"Расстояние между заводом и рабочим: {distance:.2f}")
```

```python
from abc import ABC, abstractmethod

class Calculation(ABC):
    @abstractmethod
    def calculate(self, n):
        pass

class Factorial(Calculation):
    def calculate(self, n):
        if n < 0:
            return "Факториал не определен для отрицательных чисел"
        if n == 0:
            return 1
        
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

class DoubleFactorial(Calculation):
    def calculate(self, n):
        if n < 0:
            return "Двойной факториал не определен для отрицательных чисел"
        if n == 0:
            return 1
            
        result = 1
        for i in range(n, 0, -2):
            result *= i
        return result

number_to_check = 9

print(f"Проверка для числа: {number_to_check}")

fact_calculator = Factorial()
fact_result = fact_calculator.calculate(number_to_check)
print(f"Факториал числа {number_to_check}! = {fact_result}")

double_fact_calculator = DoubleFactorial()
double_fact_result = double_fact_calculator.calculate(number_to_check)
print(f"Двойной факториал числа {number_to_check}!! = {double_fact_result}")
```

```python
class Donor:
    def __init__(self, recipient_name, recipient_blood_type, donor_name, donor_blood_type):
        self.recipient_name = recipient_name
        self.donor_name = donor_name
        
        self.recipient_blood_type = self._normalize_blood_type(recipient_blood_type)
        self.donor_blood_type = self._normalize_blood_type(donor_blood_type)

    def _normalize_blood_type(self, blood_type):
        type_map = {
            "1": 1, "i": 1,
            "2": 2, "ii": 2,
            "3": 3, "iii": 3,
            "4": 4, "iv": 4
        }
        str_type = str(blood_type).lower()
        return type_map.get(str_type, int(blood_type) if str(blood_type).isdigit() else None)

    def can_donate(self):
        print(f"Проверка: Реципиент {self.recipient_name} ({self.recipient_blood_type} гр.), "
              f"Донор {self.donor_name} ({self.donor_blood_type} гр.)")
        
        if self.recipient_blood_type is None or self.donor_blood_type is None:
            print("-> Ошибка: одна из групп крови указана неверно.")
            return False

        if self.recipient_blood_type == self.donor_blood_type:
            print("-> Результат: Да, донорство возможно.")
            return True
        else:
            print("-> Результат: Нет, группы крови не совпадают.")
            return False

case1 = Donor("Иванов И.И.", 2, "Петров П.П.", 2)
case1.can_donate()
print("-" * 20)

case2 = Donor("Сидоров С.С.", "III", "Михайлов М.М.", "IV")
case2.can_donate()
print("-" * 20)

case3 = Donor("Алексеев А.А.", 4, "Сергеев С.С.", "IV")
case3.can_donate()
```

```python
from abc import ABC, abstractmethod

class ClientOFD(ABC):
    def __init__(self, inn):
        self.inn = inn

    @abstractmethod
    def print_requisites(self):
        pass

class LegalEntity(ClientOFD):
    def __init__(self, name, inn, ogrn):
        super().__init__(inn)
        self.name = name
        self.ogrn = ogrn

    def print_requisites(self):
        print("--- Выписка для юридического лица ---")
        print(f"Название: {self.name}")
        print(f"ИНН: {self.inn}")
        print(f"ОГРН: {self.ogrn}")
        print("-" * 35)

class IndividualEntrepreneur(ClientOFD):
    def __init__(self, surname, inn, ogrnip):
        super().__init__(inn)
        self.surname = surname
        self.ogrnip = ogrnip

    def print_requisites(self):
        print("--- Выписка для ИП ---")
        print(f"Фамилия: {self.surname}")
        print(f"ИНН: {self.inn}")
        print(f"ОГРНИП: {self.ogrnip}")
        print("-" * 24)

clients = [
    LegalEntity("ООО 'Ромашка'", "7705123456", "1027700123456"),
    IndividualEntrepreneur("Иванов И.И.", "500198765432", "304500112345678"),
    LegalEntity("АО 'Технопром'", "7708654321", "1037700654321"),
    IndividualEntrepreneur("Петрова А.В.", "631912345678", "310631998765432")
]

print("Генерация выписок для всех клиентов ОФД:\n")
for client in clients:
    client.print_requisites()
```
